Ejercicio 6: Pendiente de una Rampa de Acceso

**PLANTEAMIENTO**
Se desea calcular la pendiente de una rampa de acceso, la cual es importante para garantizar que cumpla con normas de accesibilidad y seguridad. La pendiente representa qué tan inclinada es la rampa y se obtiene a partir de la relación entre la altura que se eleva y la distancia horizontal recorrida.

**REQUERIMIENTOS DEL PROBLEMA**

Entradas:
Altura de la rampa (en metros)
Longitud horizontal de la rampa (en metros)
Proceso:
Calcular la pendiente utilizando la relación entre la altura y la longitud horizontal.
La pendiente se puede expresar como una fracción, decimal o porcentaje.

𝑚
=
altura
longitud horizontal
m=
longitud horizontal
altura
	​


Salida:
Valor de la pendiente de la rampa (en decimal o porcentaje)

```python
# Entradas
altura = float(input("Ingrese la altura de la rampa (m): "))
longitud = float(input("Ingrese la longitud horizontal (m): "))

# Proceso
pendiente = altura / longitud

# Salidas
print("La pendiente de la rampa es:", pendiente)
print("La pendiente en porcentaje es:", pendiente * 100, "%")
```

Ejercicio 2: # **Dosificación de Fertilizante Líquido**

**Problema**: Un sistema de riego por goteo requiere aplicar una dosis de fertilizante por hectárea. El usuario ingresa las hectáreas totales y la dosis requerida (litros/hectárea).

**Datos**: El fertilizante se diluye en agua en una relación 1:500 (1 litro de producto por 500 de agua).

**Prueba**: 5 hectáreas, Dosis 2 L/ha. Salida: 10L de producto y 5000L de agua total.

```python
# Entradas
hectareas = float(input("Ingrese las hectáreas: "))
dosis = float(input("Ingrese la dosis (L/ha): "))

# Proceso
producto = hectareas * dosis
agua = producto * 500  # relación 1:500

# Salidas
print("Cantidad de fertilizante (L):", producto)
print("Cantidad de agua (L):", agua)
```

# **Ejercicio 3: Rendimiento de Pulpa en Frutas**

**Problema:** Una planta procesadora necesita saber el porcentaje de aprovechamiento. Se ingresa el peso total de la fruta comprada y el peso de los residuos (cáscaras y semillas).

**Datos:** Pulpa=Peso Total−Residuos
. %Rendimiento=(Pulpa/Peso Total)×100
.

**Prueba:** Peso Total = 500kg, Residuos = 150kg. **Salida:** 350kg de pulpa (70%).



```python
# Entradas
peso_total = float(input("Ingrese el peso total de la fruta (kg): "))
residuos = float(input("Ingrese el peso de los residuos (kg): "))

# Proceso
pulpa = peso_total - residuos
rendimiento = (pulpa / peso_total) * 100

# Salidas
print("Cantidad de pulpa (kg):", pulpa)
print("Rendimiento (%):", rendimiento)
```

# **Ejercicio 5: Peso de una Viga de Acero**

**Problema:** Un ingeniero necesita calcular el peso propio de una viga de sección rectangular para el análisis estructural. Se ingresa base, altura y longitud en metros.

**Datos:** Densidad del acero =7850 kg/m3
. Peso=Base×Altura×Longitud×Densidad
.

**Prueba:** Base = 0.3m, Altura = 0.5m, Largo = 6m. Salida: 7065 kg.

```python
# Entradas
base = float(input("Ingrese la base de la viga (m): "))
altura = float(input("Ingrese la altura de la viga (m): "))
longitud = float(input("Ingrese la longitud de la viga (m): "))

# Dato constante
densidad = 7850  # kg/m³

# Proceso
peso = base * altura * longitud * densidad

# Salidas
print("Peso de la viga (kg):", peso)
```

#**Ejercicio 6: Pendiente de una Rampa de Acceso**

**Problema:** Determinar si una rampa cumple con la normativa de accesibilidad. El usuario ingresa la altura a vencer y la longitud horizontal de la rampa (en metros).

**Datos:** %Pendiente=(Altura/Longitud)×100
.

**Prueba:** Altura = 0.6m, Longitud = 5m. **Salida:** 12% de pendiente.

```python
# Entradas
altura = float(input("Ingrese la altura de la rampa (m): "))
longitud = float(input("Ingrese la longitud horizontal (m): "))

# Proceso
pendiente = (altura / longitud) * 100

# Salidas
print("Pendiente de la rampa (%):", pendiente)
```
