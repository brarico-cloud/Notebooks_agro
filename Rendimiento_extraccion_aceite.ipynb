# **Control de Rendimiento en Extracción de Aceite**

Este ejercicio es ideal para introducir float, int y la concatenación de strings.

**Planteamiento**:

Una planta procesadora de palma de aceite necesita **calcular el Índice de Rendimiento de Extracción (IRE). El operario debe ingresar el peso total de los racimos de fruta fresca (RFF)** procesados en kilogramos y la cantidad de aceite crudo obtenido en litros.

**Requerimientos**:

1. Solicitar el nombre del lote de producción (Tipo string).
2. Solicitar el peso de los racimos en kg (Tipo float). (PR)
3. Solicitar los litros de aceite obtenidos (Tipo float).
4. Calcular el rendimiento sabiendo que la densidad aproximada del aceite es de 0.92 kg/L. (r)

$$PA (kg)=LT * 0.92$$

$$r=(PA/PR) * 100$$

**Prueba del programa**:

Si se ingresan 1000 kg de racimos y se obtienen 200 litros de aceite, el programa debe informar que el rendimiento del lote es del 18.4%.


````python
# ENTRADA Solicitar el nombre del lote de producción (Tipo string).
# Solicitar el peso de los racimos en kg (Tipo float). (PR)
# Solicitar los litros de aceite obtenidos (Tipo float).
nombre = input("Ingrese el nombre del lote de producción: ")
PR = float(input("Ingrese el peso de los racimos en kg: "))


# PROCESOS: Calcualar el peso del aceite en (PA) y el rendimiento (r)
LT = float(input("Ingrese la cantidad de aceite crudo obtenido en litros: "))
PA = LT * 0.92
r = (PA/PR) * 100

# SALIDA: Informar el rendimiento del lote
print("El rendimiento del lote", nombre, "es del", r, "%")
````

```markdown
Aquí está la explicación línea por línea del código proporcionado:

*   `nombre = input("Ingrese el nombre del lote de producción: ")`:
    *   Esta línea solicita al usuario que ingrese el nombre del lote de producción. La función `input()` lee la entrada del teclado como una cadena de texto y la asigna a la variable `nombre`.

*   `PR = float(input("Ingrese el peso de los racimos en kg: "))`:
    *   Similar a la anterior, esta línea solicita al usuario el peso de los racimos en kilogramos. La entrada se convierte a un número de punto flotante (`float`) mediante la función `float()` y se almacena en la variable `PR` (Peso de Racimos).

*   `LT = float(input("Ingrese la cantidad de aceite crudo obtenido en litros: "))`:
    *   Esta línea solicita al usuario la cantidad de aceite crudo obtenida en litros. La entrada también se convierte a un número de punto flotante y se guarda en la variable `LT` (Litros de Aceite).

*   `PA = LT * 0.92`:
    *   Esta línea calcula el peso del aceite (PA) en kilogramos. Multiplica la cantidad de litros de aceite (`LT`) por la densidad aproximada del aceite (0.92 kg/L). El resultado se almacena en `PA`.

*   `r = (PA/PR) * 100`:
    *   Esta línea calcula el rendimiento de extracción (`r`). Divide el peso del aceite (`PA`) por el peso de los racimos (`PR`) y luego multiplica el resultado por 100 para obtener un porcentaje. El valor se guarda en la variable `r`.

*   `print("El rendimiento del lote", nombre, "es del", r, "%")`:
    *   Finalmente, esta línea imprime el resultado en la consola. Utiliza la función `print()` para mostrar una cadena de texto que incluye el nombre del lote, el rendimiento calculado (`r`), y el símbolo de porcentaje. Los diferentes elementos se concatenan automáticamente en la salida.
```
