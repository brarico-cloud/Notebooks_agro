## ¿Qué es Python?

En términos sencillos, Python es un lenguaje de programación de **alto nivel**, lo que significa que su sintaxis es muy parecida al lenguaje humano (específicamente al inglés). Es famoso por ser "legible": leer un código en Python es casi como leer instrucciones en un manual.

Es un lenguaje **versátil y gratuito**, respaldado por una comunidad gigante que crea "librerías" (herramientas listas para usar) para ciencia, datos y control de maquinaria.

## ¿Por qué es importante en Ingeniería Agroindustrial?

La agroindustria hoy en día no es solo mover materia prima; es gestionar información. Python es vital por tres razones clave:

1. **Análisis de Datos:** Permite procesar miles de datos de cosechas, suelos o ventas para predecir rendimientos.

3. **Optimización de Procesos:** Ayuda a calcular las mezclas exactas para alimentos balanceados o las rutas de logística más cortas para evitar que los productos perecederos se dañen.

## 3 Ejemplos Sencillos en la Agroindustria

Aquí tienes tres scripts básicos que ilustran cómo Python facilita el trabajo diario:

### 1. Cálculo de Rendimiento de Extracción

Este código ayuda a calcular qué porcentaje de producto útil se obtuvo tras un proceso de transformación (como obtener aceite de palma o jugo de naranja).

```
# Cálculo de rendimiento agroindustrial
materia_prima = float(input("Ingrese kg de materia prima: "))
producto_final = float(input("Ingrese kg de producto obtenido: "))

rendimiento = (producto_final / materia_prima) * 100

print(f"El rendimiento del proceso es del {rendimiento}%")
```

### 2. Clasificación Automática por Peso (Calibrado)

Imagina un sensor en una banda transportadora. Este código clasifica la fruta según su peso para decidir si va a exportación o mercado local.

```
peso_fruta = 150  # Peso en gramos detectado por el sensor

if peso_fruta >= 200:
    categoria = "Exportación (Extra)"
elif peso_fruta >= 150:
    categoria = "Nacional (Primera)"
else:
    categoria = "Procesamiento Industrial"

print(f"La fruta pesa {peso_fruta}g y clasifica como: {categoria}")
```

### 3. Estimación de Riego según Humedad

Un ejemplo básico de cómo Python puede tomar decisiones lógicas basadas en datos de sensores de suelo.

```
humedad_suelo = 15  # Porcentaje de humedad detectado

if humedad_suelo < 20:
    print("ALERTA: Humedad baja. Activando sistema de riego...")
else:
    print("Nivel de humedad óptimo. Sistema de riego apagado.")
```
